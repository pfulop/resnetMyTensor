```python
x = tf.placeholder(tf.float32, shape=(None, 243, 243, 3))
```

```python
y = model(x)
```

```python
labels = tf.placeholder(tf.float32, shape=(None, n_classes), name="labels")
```

```python
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=y, labels=labels))
```

```python
optimizer = tf.train.AdamOptimizer(0.01).minimize(loss)
```

```python
predictions = tf.nn.softmax(y)
```

```python
correct_pred = tf.equal(tf.argmax(y, 1), tf.argmax(labels, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

```python
train_batches_per_epoch = int(np.floor(train_inputs.size / 10))
```

```python
sess.run(tf.global_variables_initializer())
```

```python
next_train = iterator.get_next()
```

```python
print(y)
```

```python
for _ in range(10):
    sess.run(iterator.initializer)
    for step in range(train_batches_per_epoch):
        batch_data, batch_labels = sess.run(next_train)
        feed_dict = {x: batch_data, labels: batch_labels}
        _, l, p = sess.run([optimizer, loss, accuracy],
                                                     feed_dict=feed_dict)
    
    print(l, p)    

```
